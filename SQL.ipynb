# Insights:
1. What is the most expensive item your company sells? The least expensive?

```python
SELECT top 1 pp.product_name
	,max(so.list_price) AS list_price
FROM sales.order_items AS so
LEFT JOIN production.products AS pp
ON so.item_id=pp.product_id
group by pp.product_name
```

```python
SELECT top 1 pp.product_name
	,min(so.list_price) as list_price
FROM sales.order_items AS so
LEFT JOIN production.products AS pp
ON so.item_id=pp.product_id
group by pp.product_name
```

2. how many different products of each category the company sells?

```python
SELECT Pc.category_name, COUNT(DISTINCT pP.product_name) as CountProducts
FROM production.products pp
LEFT JOIN production.categories Pc ON pP.category_id = Pc.category_id
GROUP BY Pc.category_name;
```

3. the top three categories with the highest average list price

```python
select top 3 pc.category_name,avg(so.list_price)AS average_list_price
from sales.order_items so
left join production.products pp
on so.product_id=pp.product_id
left join production.categories pc
on pp.category_id=pc.category_id
group by pc.category_name
order by avg(so.list_price) DESC
```

4. The average price of each brand.

```python
SELECT pb.brand_name, AVG(pp.list_price) as average_list_price
FROM production.products pp
JOIN production.brands pb ON pp.brand_id = pb.brand_id
GROUP BY pb.brand_name;
```
